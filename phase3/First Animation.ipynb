```python
import matplotlib.pyplot as plt
import numpy as np
import math
from scipy.constants import e, epsilon_0

import matplotlib.animation as animation
```

```python
fig, ax = plt.subplots()

charge1 = [2, 1, e]
electron = [0, 0, -e]
charges = [charge1]

scat_charges = ax.scatter(charge1[0], charge1[1], label="Charges")
scat_electron = ax.scatter(electron[0], electron[1], label="Electron")
ax.legend()

def get_force(charge, electron):
    r = math.sqrt((charge[0] - electron[0]) ** 2 + (charge[1] - electron[1]) ** 2)
    return electron[2] * charge[2] / 4 / math.pi / epsilon_0 / (r ** 3) * np.array([charge[0] - electron[0], charge[1] - electron[1]])

def update(frame):
    force = [0, 0]
    for charge in charges:
        force += get_force(charge, electron)
    electron[0] += force[0]
    electron[1] += force[1]
    return scat_electron
    
ani = animation.FuncAnimation(fig=fig, func=update, frames=40, interval=30)
plt.show()
```
